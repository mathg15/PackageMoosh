```python
import moosh as ms 
import numpy as np
```

```python
Layer1 = ms.Layer(1, 1, 1993).layer()
Layer2 = ms.Layer(2, 1, 682).layer()
Layer3 = ms.Layer(4, 1, 1922).layer()
Layer4 = ms.Layer(2, 1, 22).layer()
Layer5 = ms.Layer(4, 1, 65).layer()

Lay = [Layer1, Layer2, Layer3, Layer4, Layer5]
structure = ms.Structure(
    Lay[0],
    Lay[1],
    Lay[2],
    Lay[3],
    Lay[4]
).structure()

ms.Angular(structure, 1, 600).angular()
```

```python
ms.Spectrum(structure, 1, 30).spectrum()
```

```python
ms.Beam(structure, 0, 45, 600, 0.5).beam()
```

```python
Eps = np.array([1, 2, 4])
Structure = ms.RandomStructure(5, Eps).random_structure()
print(Structure)
```

```python
ms.Angular(Structure, 1, 600).angular()
```

```python
ms.Beam(Structure, 0, 20, 600, 0.5).beam()
```
